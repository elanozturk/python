```python
import pandas as pd
import numpy as np

df = pd.read_csv("dev.csv")
df.head(3)
```

```python
df=df[["Year","Country","GDP","PopTotal"]]
df
```

```python
df["perGDP"] =df["GDP"]/df["PopTotal"]
df
```

```python
df=df.sort_values(by="perGDP",ascending=False)
df
```

```python
filtre=df["Year"]==2020
df2020 = df[filtre]
```

```python
df2020
```

```python
df2020=df2020.reset_index(drop=True)
```

```python
filtre2=df2020["Country"]=="Turkiye"
df2020[filtre2]
```

```python
tablo = pd.DataFrame()


for i in range(1973,2022):
  filtre=df["Year"]==i
  dfyil = df[filtre]
  dfyil=dfyil.reset_index()
  filtre2=dfyil["Country"]=="Turkiye"
  dfyil=dfyil[filtre2]
  dfyil = dfyil.reset_index()
  dfyil = dfyil[["Country","Year","level_0"]]
  dfyil["level_0"] = dfyil["level_0"] + 1
  tablo=tablo.append(dfyil)

tablo=tablo.reset_index(drop=True)
tablo
```

```python
ulkeler = list(df["Country"].unique())
ulkeler
```

```python
tablo = pd.DataFrame()

for ulke in ulkeler:
  for i in range(1973,2022):
    filtre=df["Year"]==i
    dfyil = df[filtre]
    dfyil=dfyil.reset_index()
    filtre2=dfyil["Country"]==ulke
    dfyil=dfyil[filtre2]
    dfyil = dfyil.reset_index()
    dfyil = dfyil[["Country","Year","level_0"]]
    dfyil["level_0"] = dfyil["level_0"] + 1
    tablo=tablo.append(dfyil)

tablo=tablo.reset_index(drop=True)
tablo

```
