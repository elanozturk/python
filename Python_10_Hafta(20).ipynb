```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("Udemy.csv")
df.head(3)
```

```python
df["title"].str.contains("Python")
```

```python
def pythonbul(veri):
  if "Python" in veri:
    return True
  else:
    return False

filtre = df["title"].apply(pythonbul)
df[filtre]
```

```python
def kategoribul(veri):

  diller=["python","javascript","css","html","java",
        "c#"," c ","c++","swift","php","go"]

  for dil in diller:
    if dil in veri.lower():
      return dil

df["kategori"]=(df["title"]+" "+df["description"]).apply(kategoribul)
```

```python
df.head(10)
```

```python
ksay = df["kategori"].value_counts()
ksay
```

```python
plt.bar(ksay.index, ksay.values, color=["lightblue","deeppink"])
plt.xticks(rotation=90)
plt.show()
```

```python
hocalar = df["instructor"].value_counts()
plt.barh(hocalar.index[0:10],hocalar.values[0:10])
plt.xticks(rotation=90)
plt.show()
```

```python
#MAKİNE ÖĞRENMESİ

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

df=pd.DataFrame()
df["Büyüklük"] = [60,65,75,80,90,120,140]
df["Fiyat"] = [13000,14000,16000,17000,19000,25000,29000]
df

```

```python
# 100 m2 bir evin kirası ne kadardır tahmin edin?
x = df[["Büyüklük"]]   #aradığımız değeri etkileyen bağımsız değişkenler
y = df["Fiyat"]         #tahmin edeceğimiz değer
lr = LinearRegression()   #Model tipi
model = lr.fit(x,y)     #Öğren
model.predict([[100]])
```

```python
model.score(x,y)
```

```python
model.coef_    #katsayılar
```

```python
model.intercept_ #sabit
```

```python
plt.scatter(df["Büyüklük"],df["Fiyat"])
plt.plot(df["Büyüklük"],df["Fiyat"],color="red")
plt.show()
```

```python
model.predict([[200]])
```
