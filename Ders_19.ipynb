```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("Udemy.csv")
#df.head(3)
df.info()
```

```python
df.head(3)
```

```python
df["title"].str.contains("Python")
```

```python
def pythonbul(veri):
  if "Python" in veri:
    return True
  else:
    return False
filtre = df["title"].apply(pythonbul)
df[filtre]
```

```python
def kategoribul(veri):
  diller = ["python","javascript","css","c#","c++","html"," c ",
            "c++","java","php","go","react","angular","swift"]
  for dil in diller:
    if dil in veri.lower():
      return dil
df["kategori"] = (df["title"]+" "+df["description"]).apply(kategoribul)
```

```python
df.head(10)
```

```python
ksay = df["kategori"].value_counts()
ksay
```

```python
plt.bar(ksay.index, ksay.values,color=["lightblue","deeppink"])
plt.style.use("tableau-colorblind10")
#plt.figure(figsize=(16,5))
plt.xticks(rotation=70)
plt.show()
```

```python
hocalar = df["instructor"].value_counts()
plt.barh(hocalar.index[0:10],hocalar.values[0:10])
#plt.xticks(rotation=90)
plt.show()
```
