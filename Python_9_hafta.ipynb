```python
import pandas as pd
import numpy as np

sampi = pd.read_csv("tcl.csv",delimiter=";")
sampi.head(3)
```

```python
filtre=sampi["points"]==sampi["puanrevize"]
degisim = sampi[filtre]
degisim = degisim.sort_values(by="year")
degisim=degisim.reset_index(drop=True)    #index değerlerini sıfırlar
degisim['year'][0]
```

```python
link=("https://en.wikipedia.org/wiki/List_of_countries_by_vehicles_per_capita")

arac = pd.read_html(link)[0]
arac.head(3)
```

```python
arac= arac[['Ülke','1000 kişiye düşen araç','Notlar']] #istediğimiz sütunları seçme
```

```python
arac.columns=["Ülke","Adet","Araç Sayısı","Yıl"]   #Sütun isimlerini düzenleme
```

```python
arac.head()
```

```python
#arac["Yıl"].str.split("[")
#arac["Yıl"].str.split("[",expand=True)
#arac["Yıl"].str.split("[",expand=True)[0]
arac["Yıl"] = arac["Yıl"].str.split("[",expand=True)[0]
```

```python
arac.head()
```

```python
arac.info()
```

```python
arac['Araç Sayısı'] = arac['Araç Sayısı'].str.split("[",expand=True)[0]
```

```python
arac['Araç Sayısı']=arac['Araç Sayısı'].str.replace(",","")
```

```python
arac['Araç Sayısı'] = pd.to_numeric(arac['Araç Sayısı'])
```

```python
arac["nüfus"]=arac["Araç Sayısı"]/arac["Adet"]*1000
arac.head()
```

```python
arac["nüfus"]=arac["nüfus"].round()
```
