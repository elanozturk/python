```python
import numpy as np   #matematiksel
import pandas as pd   #tablolar ve veri okuma, yazma
import matplotlib.pyplot as plt   #veri görselleştirme
from sklearn.linear_model import LinearRegression   #Lineer Regresson
```

```python
df = pd.read_csv("score_updated.csv")
df.head(3)
```

```python
plt.figure(figsize=(16,6))
plt.scatter(df['Hours'],df['Scores'])
plt.xlabel("Çalışma Saati")
plt.ylabel("Sınav Puanı")
plt.show()
```

```python
x = df[['Hours']]
y = df['Scores']
lr = LinearRegression()
model = lr.fit(x,y)
```

```python
model.predict([[8]])
```

```python
model.score(x,y)
```

```python
model.predict(df[['Hours']])
```

```python
df['tahmin'] = model.predict(df[['Hours']])
```

```python
df
```

```python
df["sapma"] = df["Scores"] -df["tahmin"]
df.head(3)
```

```python
df['oran'] = (1-abs(df['sapma'])/df['Scores'])*100
df
```

```python
df['oran'].mean()
```

```python
import seaborn as sns
df = sns.load_dataset("titanic")
```

```python
df.head(3)
```

```python
df = df[["survived","pclass","age","sibsp","parch","fare","embark_town","who"]]
df.head(3)
```

```python
df['who'].unique()
```

```python
df = pd.get_dummies(df,columns=['embark_town',"who"],drop_first=True)
df.head(3)
```

```python
df
```

```python
df=df.dropna()   #Sütunlarda boş değer var ise bu satırları siler
```

```python
df
```

```python
x = df.drop("survived",axis=1)
y = df["survived"]
x_train, x_test, y_train, y_test = train_test_split(x,y,train_size=0.75,random_state=44)
tree = DecisionTreeClassifier()
model = tree.fit(x_train,y_train)
```

```python
df.head(3)
```

```python
model.predict([[2,19,0,4,30,1,0,0,1]])
```

```python
model.predict([[3,49,0,0,7,0,1,1,0]])
```

```python
model.score(x_test,y_test)
```
