```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Verileri oluştur
büyüklük = np.array([60, 65, 75, 80, 90, 120, 140]).reshape(-1, 1)
fiyat = np.array([13000, 14000, 16000, 17000, 19000, 25000, 29000])

# Modeli oluştur
model = LinearRegression()
model.fit(büyüklük, fiyat)

# 100m2 lik evin kira tahmini
ev_büyüklüğü = np.array([[100]])
tahmini_kira = model.predict(ev_büyüklüğü)

print("100m2 lik evin tahmini kirası:", tahmini_kira[0])
```

```python
model.coef_ #katsayılr
```

```python
model.intercept_ #sabit
```

```python
df = pd.read_csv("arac.csv")
df
```

```python
df["model"].unique()
```

```python
df["model"] = df["model"].str.strip()
df["model"].unique()
```

```python
df.columns = ["model","yil","fiyat","vites","km","yakitturu","vergi","yakittuk","motor","marka"]
df
```

```python
dfdum = pd.get_dummies(df,columns=["model","vites","yakitturu","marka"])
```

```python
x = dfdum.drop("fiyat",axis=1)
y = dfdum["fiyat"]

lr = LinearRegression()
model = lr.fit(x,y)
model.score(x,y)
```

```python
model.predict([["A6",2017,"Automatic",15000,"Petrol",30.4,64,2,"audi"]])
```
