```python
def yarimat(n):
  a = np.eye(n)
  for i in range(1,n):
    a=a+np.eye(n,k=i)
  return a
```

```python
yarimat(7)
```

```python
star["me_lon"]=29.02373
star["me_lat"]=40.98911
star
```

```python
# mesafe hesabı = ( (x1-x2)**2 + (y1-y2)**2 )**0.5
```

```python
star["mesafe"] = ((star["longitude"]-star["me_lon"])**2 + (star["latitude"]-star["me_lat"])**2)**0.5

```

```python
star["mesafe"] = star["mesafe"]*100
```

```python
star.sort_values(by="mesafe") #sıralama metodu
```

```python
star.sort_values(by="mesafe",ascending=False) #sıralama metodu
```
